```python
import json
import time

# ----------------------------
# Decorator to measure execution time
# ----------------------------
def execution_time(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"\nExecution time: {end - start:.5f} sec")
        return result
    return wrapper

# ----------------------------
# OOP - Student class
# ----------------------------
class Student:
    def __init__(self, student_id, name, grades=[]):
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def add_grade(self, grade):
        self.grades.append(grade)

    def average_grade(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

    def display_info(self):
        print(f"ID: {self.student_id}, Name: {self.name}, Grades: {self.grades}, Avg: {self.average_grade():.2f}")

# ----------------------------
# File Handling
# ----------------------------
def save_students(students, filename="students.json"):
    data = [{"id": s.student_id, "name": s.name, "grades": s.grades} for s in students]
    with open(filename, "w") as f:
        json.dump(data, f)
    print(f"\n{len(students)} students saved to {filename}")

def load_students(filename="students.json"):
    students = []
    try:
        with open(filename, "r") as f:
            data = json.load(f)
            for item in data:
                students.append(Student(item["id"], item["name"], item["grades"]))
    except FileNotFoundError:
        print("File not found! Starting with empty list.")
    return students

# ----------------------------
# Generator for high scorers
# ----------------------------
def high_scorers(students, threshold=80):
    for student in students:
        if student.average_grade() >= threshold:
            yield student

# ----------------------------
# Recursion example - factorial
# ----------------------------
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n-1)

# ----------------------------
# Main Program
# ----------------------------
@execution_time
def main():
    students = load_students()

    # Add new students
    s1 = Student(101, "Rajkumar", [85, 90, 78])
    s2 = Student(102, "Sagar", [65, 70, 72])
    students.extend([s1, s2])

    # Display all students
    print("\nAll Students:")
    for s in students:
        s.display_info()

    # Save students
    save_students(students)

    # Show high scorers
    print("\nHigh Scorers (>80):")
    for s in high_scorers(students):
        s.display_info()

    # Factorial
    print("\nFactorial of 5:", factorial(5))

# Run main
if __name__ == "__main__":
    main()
```
